## 1.0 Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split

%matplotlib inline
```

## 2.0 Load the dataset

```python
df = pd.read_csv('data/WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

## 3.0 Rename columns and Normalize string values

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
string_columns = list(df.dtypes[df.dtypes == 'object'].index)
for col in string_columns:
    df[col] = df[col].str.lower().str.replace(' ', '_')
```

## 4.0 Fix dtypes

```python
total_charges = pd.to_numeric(df['totalcharges'], errors='coerce')

df['totalcharges'] = pd.to_numeric(df['totalcharges'], errors='coerce')
df['totalcharges'] = df['totalcharges'].fillna(0)
```

## 5.0 Normalize class column values

```python
df['churn'] = (df['churn'] == 'yes').astype(int)
```

## 6.0 Train, validation and test split

```python
df_train_full, df_test = train_test_split(df, test_size=0.2, random_state=42)
df_train, df_val = train_test_split(df_train_full, test_size=0.33, random_state=43)
```

```python
y_train = df_train['churn'].values
y_val = df_val['churn'].values
```

```python
del df_train['churn']
del df_val['churn']
```

### 7.4 Feature importance

#### 7.4.1 Proportion Difference and Risk Ratio

```python
from IPython.display import display

global_mean = df_train_full['churn'].mean()
for col in categorical:
    df_group = df_train_full.groupby(by=col)['churn'].agg(['mean'])
    df_group['diff'] = df_group['mean'] - global_mean
    df_group['rate'] = df_group['mean'] / global_mean
    display(df_group)
```

#### 7.4.2 Mutual Information

```python
from sklearn.metrics import mutual_info_score

df_mi = df_train_full[categorical].apply(lambda x: mutual_info_score(x, df_train_full['churn']))
df_mi = df_mi.sort_values(ascending=False).to_frame(name='MI')
df_mi
```
